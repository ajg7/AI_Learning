```python
def add(*args):
  total = 0
  for i in args:
    total += i
  print(total)

add(4,3,2,5,6,7,8,9,19,2)

def multiply(*args):
  total = 1
  for i in args:
    total *= i
  print(total)
  
multiply(3,2,1,2,3,4,99,12,32)
```

```python
tasks = []

def total_list_operations(operation, value = ""):
  if operation == 1: tasks.append(value)
  elif operation == 2: tasks.pop()
  elif operation == 3: tasks[value]
  elif operation == 4: tasks.clear()
  
  return tasks

print(total_list_operations(1, "Go shopping"))
print(total_list_operations(1, "Go to the store"))
print(total_list_operations(1, "Go to friend's house"))
print(total_list_operations(2))
print(total_list_operations(3, 1))
print(total_list_operations(4))
```

```python
def calculate(num1, num2, operator):
  if operator == '+': return num1 + num2
  elif operator == '-': return num1 - num2
  elif operator == '*': return num1 * num2
  elif operator == "/": return num1 / num2

  return 0

print(calculate(9, 8, "+"))
print(calculate(9, 8, "-"))
print(calculate(9, 8, "*"))
print(calculate(9, 8, "/"))
```

```python
#Make a function to reverse a string

def reverse_string(str):
  new_str = ""
  for i in range(len(str) - 1, -1, -1):
    new_str += str[i]
  return new_str

print(reverse_string("racecar"))
```

```python
def get_even_number(nums):
  even_nums = []
  for num in nums:
    if num % 2 == 0:
      even_nums.append(num)
  return even_nums

print(get_even_number([1,2,3,4,5,6,7,8,9]))
```
